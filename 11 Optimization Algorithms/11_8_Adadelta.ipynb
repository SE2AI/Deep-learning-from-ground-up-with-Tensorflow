```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
```

### Data

```python
features, labels = make_moons(n_samples=5000, random_state=42, noise=0.1)
```

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    # The examples are read at random, in no particular order
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        j = tf.constant(indices[i: min(i + batch_size, num_examples)])
        yield tf.gather(features, j), tf.gather(labels, j)
```

```python
batch_size = 10

for X, y in data_iter(batch_size, features, labels):
    print(X, '\n', y)
    break
```

### Initializing Model Parameters

```python
num_inputs = 2
num_outputs = 2
num_hiddens = 200

W1 = tf.Variable(tf.random.normal(shape=(num_inputs, num_hiddens), mean=0, stddev=0.01))
b1 = tf.Variable(tf.zeros(num_hiddens))
W2 = tf.Variable(tf.random.normal(shape=(num_hiddens, num_hiddens), mean=0, stddev=0.01))
b2 = tf.Variable(tf.random.normal([num_hiddens], stddev=.01))
W3 = tf.Variable(tf.random.normal(shape=(num_hiddens, num_outputs), mean=0, stddev=0.01))
b3 = tf.Variable(tf.random.normal([num_outputs], stddev=.01))

params = [W1, b1, W2, b2, W3, b3]
```

### Softmax Activation

```python
def softmax(X):
    X_exp = tf.exp(X)
    partition = tf.reduce_sum(X_exp, 1, keepdims=True)
    return X_exp / partition  # The broadcasting mechanism is applied here
```

### ReLU Activation

```python
def relu(X):
    return tf.math.maximum(X, 0)
```

### Defining the Model

```python
def net(X):
    X = tf.reshape(tf.cast(X, dtype=tf.float32), (-1, num_inputs))
    H1 = relu(tf.matmul(X, W1) + b1)
    H2 = relu(tf.matmul(H1, W2) + b2)
    return softmax(tf.matmul(H2, W3) + b3)
```

### Loss Function

```python
def cross_entropy(y_hat, y):
    return -tf.math.log(tf.boolean_mask(y_hat, tf.one_hot(y, depth=y_hat.shape[-1])))
```

### Optimization

#### It is referred to as not having a learning rate since it uses the amount of change itself as calibration for future change.

#### The Algorithm

$\begin{aligned}
    \mathbf{s}_t & = \rho \mathbf{s}_{t-1} + (1 - \rho) \mathbf{g}_t^2
\end{aligned}$

$\begin{split}\begin{aligned}
    \mathbf{x}_t  & = \mathbf{x}_{t-1} - \mathbf{g}_t'\\
\end{aligned}\end{split}$

$\begin{split}\begin{aligned}
    \mathbf{g}_t' & = \frac{\sqrt{\Delta\mathbf{x}_{t-1} + \epsilon}}{\sqrt{{\mathbf{s}_t + \epsilon}}} \odot \mathbf{g}_t \\
\end{aligned}\end{split}$

$\begin{aligned}
    \Delta \mathbf{x}_t & = \rho \Delta\mathbf{x}_{t-1} + (1 - \rho) {\mathbf{g}_t'}^2
\end{aligned}$

```python
def init_adadelta_states():
    s_W1 = tf.Variable(tf.zeros((num_inputs, num_hiddens)))
    s_b1 = tf.Variable(tf.zeros(num_hiddens))
    s_W2 = tf.Variable(tf.zeros((num_hiddens, num_hiddens)))
    s_b2 = tf.Variable(tf.zeros([num_hiddens]))
    s_W3 = tf.Variable(tf.zeros((num_hiddens, num_outputs)))
    s_b3 = tf.Variable(tf.zeros(num_outputs))
    
    delta_W1 = tf.Variable(tf.zeros((num_inputs, num_hiddens)))
    delta_b1 = tf.Variable(tf.zeros(num_hiddens))
    delta_W2 = tf.Variable(tf.zeros((num_hiddens, num_hiddens)))
    delta_b2 = tf.Variable(tf.zeros([num_hiddens]))
    delta_W3 = tf.Variable(tf.zeros((num_hiddens, num_outputs)))
    delta_b3 = tf.Variable(tf.zeros(num_outputs))
    
    return ((s_W1, delta_W1), (s_b1, delta_b1), (s_W2, delta_W2), (s_b2, delta_b2), (s_W3, delta_W3), (s_b3, delta_b3))
```

```python
def minibatch_sgd_adadelta(params, grads, states, rho):
    eps = 1e-5
    for p, (s, delta), grad in zip(params, states, grads):
        s[:].assign(rho * s + (1 - rho) * tf.math.square(grad))
        g = (tf.math.sqrt(delta + eps) / tf.math.sqrt(s + eps)) * grad
        p[:].assign(p - g)
        delta[:].assign(rho * delta + (1 - rho) * g * g)
```

### Training

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
num_epochs = 10
states = init_adadelta_states()
rho = 0.9
loss = cross_entropy
```

```python
for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, X_train, y_train):
      with tf.GradientTape() as g:
        l = tf.math.reduce_mean(loss(net(X), y))

      grads = g.gradient(l, params)
      minibatch_sgd_adadelta(params, grads, states, rho)
```

### Decision Boundary

```python
def plot_decision_boundary(net, X, y):
    # Define the axis boundaries of the plot and create a meshgrid
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                       np.linspace(y_min, y_max, 100))
    # Create X values (we're going to predict on all of these)
    x_in = np.c_[xx.ravel(), yy.ravel()] 
    # Make predictions using the trained model
    y_pred = np.array(net(x_in)).argmax(axis=1)
    # Check for multi-class
    if y_pred[0].size > 1:
        print("doing multiclass classification...")
        # We have to reshape our predictions to get them ready for plotting
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print("doing binary classifcation...")
        y_pred = np.round(y_pred).reshape(xx.shape)
    # Plot decision boundary
    plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```

```python
plot_decision_boundary(net, X_test, y_test)
```
